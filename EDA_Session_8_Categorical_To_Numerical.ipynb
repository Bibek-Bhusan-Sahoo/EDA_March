```python
visa_df = pd.read_csv(r'C:\Users\LEGION\OneDrive\Documents\Naresh IT\Datafiles\Visadataset.csv')
for column in visa_df.select_dtypes(include='object'):
    uniqueLabels = list(visa_df[column].unique())
    numericalColVal = [i for i in range(len(uniqueLabels))]
    dict1 = dict(zip(uniqueLabels, numericalColVal))
    visa_df[column] = visa_df[column].map(dict1)
visa_df
```

**np.where()**

- np.where is applicable only for binary values
- np.where takes three arguments
    - condition
    - value1 if True 
    - value2 if False

$Scenario$

- For example if case_status value equal to certified replace with 0
- otherwise replace with 1

```python
visa_df = pd.read_csv(r'C:\Users\LEGION\OneDrive\Documents\Naresh IT\Datafiles\Visadataset.csv')
con1 = visa_df['case_status'] == "Certified"

visa_df['case_status']= np.where(con1, 1, 0)
visa_df
```

**One hot encoder**

- One hot means if one is ON another one is OFF
- ON represents with 1, and OFF represents with 0
- For example: case status has two unique lables
    - Denied
    - Certified
- If you apply One hot encoder on case_status column it will create two extra columns
    - case_status_Denied
    - case_status_Certified

|case_status|case_status_Denied|case_status_Certified|
|--|--|--|
|Denied|1|0
|Certified|0|1

**Advantages**

- In Machine Learning we have one important property is **Indepenedent of variables**
- Machine Learning models expects that variables are **independent of each other**
- One hot encoder creating new columns, the new columns are **Independent of each other**
- Independent variables means both variables have **90 degree phase shift**
- 90 degree phase shift also called as **Perpendicular each other**
- Perpendicular means **Orthogonal each other**

**Disadvantage**

- One hot encoder creates new columns equal to number of unique lables.
- If number of unique labels are increased, new columns also increased.
- The more column leads to data is more
- When data is more, then complexity will increase
- Complexity increases more computation power is required
- Also time also increased
- The enitre process is called as **curse of dimensionality**.
