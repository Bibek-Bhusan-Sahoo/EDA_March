**Data Transformations**

- Generally data has three types of distributions:-
    - Normal Distribution
    - Positive Scew
    - Negative Scew
    
- All the math concepts were developed by assumption as data follows Normal Distribution
- Data scew happens because of many reasons, in that one reason is outliers
- Once we perform the outlier analysis, then we can apply transformation techniques, if the data does not follows normality
- We have some transformation techniques mentioned below:-
    - Log Transformation
    - Exponential Transformation
    - Reciprocal Transformations
    - Square root Transformations
    - Power Transformation

$Step-1$

```python
import numpy as np
import matplotlib.pyplot as plt
```

$Step-2$

**Read the data**

```python
exp_data = np.random.exponential(size=10000)
exp_data
```

$Step-3$

**Plot the Data**

```python
plt.hist(exp_data, bins=20, label="Exponential")
plt.legend()
plt.show()
```

```python
norm_data = np.random.normal(size=10000)
norm_data
```

```python
plt.hist(norm_data, bins=20, label="Normal Data")
plt.legend()
plt.show()
```

**The Goal is to convert exp_data to Normal Distribution**

- bins = intervals

$Step-4$

**Log Transformation**

- Log Transformation means performing logarithm operations on original data
- It is an approach to convert data into Normal Distribution
- Log means Natural Logarithm base = e

```python
x = 2
np.log(x)
```

```python
log_data = np.log(exp_data)
plt.hist(log_data, bins=20, label="Logarithm Data")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 3))
plt.suptitle("Exponential Data vs Logarithm Transformation Data")
plt.subplot(1, 2, 1).hist(exp_data, bins=20, label="Exponential")
plt.legend()
log_data = np.log(exp_data)
plt.subplot(1, 2, 2).hist(log_data, bins=20, label="Logarithm Data")
plt.legend()
plt.show()
```

**Reciprocal Transformation**

- Reciprocal will fail when values are very small
- For Example: x = 0 then 1/0 will fail

```python
x = 2
1/x
```

```python
reciprocal_data = np.reciprocal(exp_data)
reciprocal_data
```

```python
log_data = np.reciprocal(exp_data)
plt.hist(log_data, bins=20, label="Reciprocal Data")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 3))
plt.suptitle("Exponential Data vs Logarithm Transformation Data")
plt.subplot(1, 2, 1).hist(exp_data, bins=20, label="Exponential")
plt.legend()
reciprocal_data = np.reciprocal(exp_data)
plt.subplot(1, 2, 2).hist(reciprocal_data, bins=20, label="Reciprocal Data")
plt.legend()
plt.show()
```

**Square Root Transformation**

```python
import math
x = 2
math.sqrt(x)
```

```python
square_root_data = np.sqrt(exp_data)
plt.hist(square_root_data, bins=20, label="Square Root Data")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(14, 5))
plt.suptitle("Exponential Data vs Logarithm Transformation Data")
plt.subplot(2, 2, 1).hist(exp_data, bins=20, label="Exponential")
plt.legend()
log_data = np.log(exp_data)
plt.subplot(2, 2, 2).hist(log_data, bins=20, label="Logarithm Data")
plt.legend()
plt.show()

reciprocal_data = np.reciprocal(exp_data)
plt.subplot(2, 2, 3).hist(reciprocal_data, bins=20, label="Reciprocal Data")
plt.legend()
plt.show()

square_root_data = np.sqrt(exp_data)
plt.subplot(2, 2, 4).hist(square_root_data, bins=20, label="Square Root Data")
plt.legend()
plt.show()
```

**Power Transformation**

- Two Ways to perform Power Transformation:-
    - Ye Johnson: It is under sklearn package
    - Box Cox:

**Box Cox**

<img src="https://i.imgur.com/Mq1kQzy.jpg" jsaction="VQAsE" class="sFlh5c pT0Scc iPVvYb" style="max-width: 436px; width: 436px; height: 123px; margin: 12.5px 0px;" alt="Box Cox transformation in R | R-bloggers" jsname="kn3ccd" aria-hidden="false">

```python
exp_data
```

```python
exp_data.reshape(-1, 1) # power transformation expects list in list
```

```python
from sklearn.preprocessing import PowerTransformer

box_cox = PowerTransformer(method="box-cox", standardize=True)
boxCox_data  = box_cox.fit_transform(exp_data.reshape(-1,1))
plt.hist(boxCox_data, bins=20, label="Box Cox Data")
plt.legend()
plt.show()
```

**Ye Johnson**

<img src="https://miro.medium.com/v2/resize:fit:1200/1*Vv67g1vBVx483FYZ1dgxQQ.png" jsaction="VQAsE" class="sFlh5c pT0Scc iPVvYb" style="max-width: 600px; width: 557px; height: 121px; margin: 13.5px 0px;" alt="Bring your data to normal distribution with imperio YeoJohnsonTransformer.  | by Vasile Păpăluță | softplus-publication | Medium" jsname="kn3ccd" aria-hidden="false">

```python
from sklearn.preprocessing import PowerTransformer

ye_johnson = PowerTransformer(method="yeo-johnson", standardize=True)
ye_johnson_data  = ye_johnson.fit_transform(exp_data.reshape(-1,1))
plt.hist(ye_johnson_data, bins=20, label="Ye-Johnson Data")
plt.legend()
plt.show()
```
