**task**

- Create categorical column list
- Create numerical column list

**idea**

- convert series type to dictionary
- iterate through loop
- apply the conditions and get result in the list

```python
columnTypes = dict(visa_df.dtypes)
categoricalColumn = []
numericalColumn = []
for key, value in columnTypes.items():
    if value == 'object':
        categoricalColumn.append(key)
    else:
        numericalColumn.append(key)
        
print(f"Categorical Columns : {categoricalColumn}", f"Numerical Column : {numericalColumn}", sep="\n")
```

**Using list comprehension**

```python
columnTypes = dict(visa_df.dtypes)
categoricalCols = [key for key, value in columnTypes.items() if value in ['object']]
numericalCols  = [key for key, value in columnTypes.items() if value not in ['object']]
print(f"Categorical Columns : {categoricalCols}", f"Numerical Column : {numericalCols}", sep="\n")
```

$take$

- It is used to extract the particular index of data
- take has one parameter called axis
- axis = 1 means columns
- axis = 0 means rows
- by default axis = 0 is there
- in below example we given 100, 200, 300 and we didn't provide axis values
- it means we are extracting rows

```python
visa_df.take([100, 200, 300])
```

```python
visa_df.take([100, 200, 300], axis=1)
```

```python
visa_df.take(range(10, 20))

# Extract rows from index 10 till 19
```

```python
visa_df.take([2, 5, 7], axis=1)
```

```python
visa_df.take(range(2, 5), axis=1)
```

```python
visa_df.take([100:105]) # Not working
```

-  I want 100,200,300 rows and 2,5,7 column in a single take() call
- with single syntax it is not possible by using **take()**

```python
visa_df.take([100, 200, 300]).take([2,5,7], axis=1)
```
